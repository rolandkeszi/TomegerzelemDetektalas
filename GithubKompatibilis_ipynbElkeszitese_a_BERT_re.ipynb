```python
# ============================================================
# GitHub-kompatibilis .ipynb készítése
# Kódok + outputok megtartása, Colab-widgetek eltávolítása
# ============================================================

import json
from pathlib import Path

# ------------------------------------------------------------
# 1. Az eredeti notebook pontos elérési útja
# ------------------------------------------------------------

input_path = Path("/content/drive/MyDrive/Privat/Study/PE/SZD/IMDb_BERT/BERT_sentiment_colab.ipynb")

# ------------------------------------------------------------
# 2. Ide készül a GitHubra feltölthető tisztított notebook
# ------------------------------------------------------------

output_path = Path("/content/drive/MyDrive/Privat/Study/PE/SZD/IMDb_BERT/BERT_sentiment_GitHub_ready.ipynb")

# ------------------------------------------------------------
# 3. Notebook betöltése
# ------------------------------------------------------------

with open(input_path, "r", encoding="utf-8") as f:
    nb = json.load(f)

# ------------------------------------------------------------
# 4. Notebook-szintű Colab/Jupyter widget metaadatok törlése
#    FONTOS: az outputs mezőket NEM töröljük!
# ------------------------------------------------------------

nb.get("metadata", {}).pop("widgets", None)

# ------------------------------------------------------------
# 5. Cellaszintű zavaró metaadatok törlése
#    A kód, markdown és output változatlanul megmarad.
# ------------------------------------------------------------

for cell in nb.get("cells", []):
    metadata = cell.get("metadata", {})

    # Colab/HuggingFace widget-hivatkozások törlése
    metadata.pop("referenced_widgets", None)

    if "colab" in metadata and isinstance(metadata["colab"], dict):
        metadata["colab"].pop("referenced_widgets", None)

    # Output ID-k opcionális törlése; az output tartalma megmarad
    metadata.pop("outputId", None)

# ------------------------------------------------------------
# 6. Execution count rendezése:
#    Ha egy cellának van outputja, de execution_count = null,
#    akkor sorszámot adunk neki. Ez GitHubon rendezettebb.
# ------------------------------------------------------------

execution_counter = 1

for cell in nb.get("cells", []):
    if cell.get("cell_type") == "code":
        outputs = cell.get("outputs", [])

        if outputs:
            cell["execution_count"] = execution_counter

            for output in outputs:
                if "execution_count" in output:
                    output["execution_count"] = execution_counter

            execution_counter += 1
        else:
            cell["execution_count"] = None

# ------------------------------------------------------------
# 7. Notebook mentése
# ------------------------------------------------------------

with open(output_path, "w", encoding="utf-8") as f:
    json.dump(nb, f, ensure_ascii=False, indent=1)

print("Elkészült a GitHubra feltölthető notebook:")
print(output_path)
```
