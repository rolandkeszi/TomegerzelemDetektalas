#Megvizsgált modellek pontossága egy chart-on

```python
import matplotlib.pyplot as plt
import pandas as pd

# Adatok előkészítése
data = {
    'Model': [
        'OpenAI gpt-5.3-chat-latest', 'Anthropic claude-sonnet4-6', 'OpenAI gpt-4.1-mini',
        'CNN (N=50.000)', 'Anthropic claude-haiku-4-5', 'Google gemini-2,5-flash-lite',
        'Google gemini-2,5-flash', 'LSTM (N=50.000)', 'SNN (N=50.000)',
        'DistilBERT, HF (N=50.000)', 'DistilBERT, HF (N=1.000)', 'OpenAI, ChatGPT - ZSP (N=1.000)',
        'NLTK - TextBlob (N=50.000)', 'NLTK - VADER (N=50.000)'
    ],
    'Accuracy': [
        96.90, 96.10, 95.80, 94.67, 94.50, 94.48, 93.37, 87.38, 86.36, 82.75, 80.90, 69.99, 68.95, 68.09
    ]
}

df = pd.DataFrame(data)

# Diagram stílusának beállítása
plt.figure(figsize=(12, 8))
colors = plt.cm.viridis([i/len(df) for i in range(len(df))][::-1]) # Színátmenet a jobb vizuális hatásért

bars = plt.bar(df['Model'], df['Accuracy'], color=colors, edgecolor='black', alpha=0.8)

# Értékek ráírása az oszlopokra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval}%', ha='center', va='bottom', fontweight='bold')

# Tengelyek és címek
plt.title('Modellek pontossági összehasonlítása', fontsize=16, fontweight='bold', pad=20)
plt.ylabel('Pontosság (%)', fontsize=12)
plt.xlabel('Modell megnevezése', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 110) # Hely az adatok felett
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Adatok az összes megadott modell alapján
data = {
    'Model': [
        'Simple neural net (N=50.000)', 'Convolutional neural net (N=50.000)', 'Recurrent neural net (N=50.000)',
        'DistilBERT, HF (N=1.000)', 'DistilBERT, HF (N=50.000)', # Most már mindkét BERT szerepel
        'Python NLTK - TextBlob (N=60.000)', 'Python NLTK - VADER (N=60.000)',
        'OpenAI, ChatGPT - ZSP (N=1.000)', 'OpenAI gpt-5.3-chat-latest', 'OpenAI gpt-4.1-mini',
        'Google gemini-2.5-flash', 'Google gemini-2.5-flash-lite',
        'Anthropic claude-haiku-4-5', 'Anthropic claude-sonnet-4-6'
    ],
    'Accuracy': [
        86.4, 94.7, 87.4,
        80.90, 82.75, # A két BERT pontossága
        69.0, 68.1,
        70.0, 96.9, 95.8,
        93.4, 94.5,
        96.2, 96.1
    ]
}

df = pd.DataFrame(data)

# Színkódok frissítése az új oszlophoz igazítva
colors = (['#D31F33'] * 3 +    # NEURAL NET
          ['#F2C994'] * 2 +    # BERT based (most már 2 elem)
          ['#5D3291'] * 2 +    # Lexikon based
          ['#3E9D84'] * 3 +    # OpenAI
          ['#5C88D1'] * 2 +    # Google
          ['#7A4B2A'] * 2)     # Anthropic

plt.figure(figsize=(16, 10))
bars = plt.bar(df['Model'], df['Accuracy'], color=colors, edgecolor='black', linewidth=0.5, width=0.7)

# Értékek ráírása
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval}%',
             ha='center', va='bottom', fontweight='bold', fontsize=9)

# Kategória feliratok pozícióinak újrakalibrálása
def add_source_label(start_idx, end_idx, label, y_pos=110):
    plt.hlines(y_pos, start_idx - 0.4, end_idx + 0.4, colors='black', linewidth=1.2)
    plt.text((start_idx + end_idx) / 2, y_pos + 1.5, label,
             ha='center', va='bottom', fontweight='bold', fontsize=10)

add_source_label(0, 2, "NEURAL NET - GloVe")
add_source_label(3, 4, "BERT based") # Kiterjesztve a 4-es indexig
add_source_label(5, 6, "Lexikon based")
add_source_label(7, 13, "LLMs - REST API: AI as a Service (AIaaS)")

# Al-csoportok (LLM)
plt.text(8, 103, "OpenAI Models", ha='center', fontsize=9, fontweight='semibold')
plt.text(10.5, 103, "Google Models", ha='center', fontsize=9, fontweight='semibold')
plt.text(12.5, 103, "Anthropic Models", ha='center', fontsize=9, fontweight='semibold')

# Esztétika
plt.ylabel('Accuracy (%)', fontsize=12, fontweight='bold')
plt.title('A megvizsgált modellek pontossági összehasonlítása (Összes model)', fontsize=16, fontweight='bold', pad=50)
plt.ylim(0, 125)
plt.xticks(rotation=40, ha='right', fontsize=9)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```
